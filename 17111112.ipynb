```python
import numpy as np
import pandas as pd
```

```python
# read the date into dataframe
df = pd.read_csv('flights.csv')
```

```python
df.head(5)
```

```python
# drop the columns that not related
df_n=df.drop(df.columns[[0,5,6,12,13,14,15,16,18,19,23,24,25,26,27,28,29,30]], axis=1, inplace=False)
```

```python
# replace the airline information with numbers in map x
df_al = pd.read_csv('airlines.csv')
airlinecode=pd.Series([x for x in range(len(df_al))])
airlmap=airlinecode.map(df_al['IATA_CODE'])
x = {v: k for k, v in airlmap.items()}
```

```python
# replace the airport information with numbers in map y
df_ap = pd.read_csv('airports.csv')
airportcode=pd.Series([x for x in range(len(df_ap))])
airlmap=airportcode.map(df_ap['IATA_CODE'])
y = {v: k for k, v in airlmap.items()}
```

```python
df_nr=df_n.replace({'AIRLINE': x,'ORIGIN_AIRPORT': y,'DESTINATION_AIRPORT': y})
df_nr.head(5)
```

```python
# take samples
df_nr_0002 = df_nr.sample(frac=0.002)
```

```python
#check the correlations among features
corr = df_nr_0002.corr()
corr
```

```python
df_nr_0002.describe()
```

```python
df_nr_0002 = df_nr_0002.dropna()
```

```python
from sklearn.decomposition import PCA
df_nr_0002.isnull().any()
```

```python
# do pca analysis on the data
pca=PCA(n_components = 6)
pcafits=pca.fit(df_nr_0002)
pcafits
```
